```python
# Day 1
data = open('data/day1.txt').read().strip().split('\n')
# Part 1
pos = 50
count = 0
for line in data:
    if line[0] == "R":
        pos += int(line[1:]) % 100
    else:
        pos -= int(line[1:]) % 100
    if pos > 99:
        pos = pos - 100
    if pos < 0:
        pos = 100 + pos
    if pos == 0:
        count += 1
print("Result 1 : ",count)

# Part 2
pos = 50
count = 0
for line in data:
    if line[0] == "R":
        ctr = 100 - pos
        if int(line[1:]) >= ctr:
            count += ((int(line[1:]) - ctr) // 100 ) + 1
        pos += int(line[1:]) % 100
    else:
        ctr = pos
        if pos == 0:
            ctr = 100
        if int(line[1:]) >= ctr:
            count += ((int(line[1:]) - ctr) // 100) + 1
        pos -= int(line[1:]) % 100
    if pos > 99:
        pos = pos - 100
    if pos < 0:
        pos = 100 + pos
print("Result 2 : ",count)
```
