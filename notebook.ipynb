```python
# Day 1
data = open('data/day1.txt').read().strip().split('\n')
# Part 1
pos = 50
count = 0
for line in data:
    if line[0] == "R":
        pos += int(line[1:]) % 100
    else:
        pos -= int(line[1:]) % 100
    if pos > 99:
        pos = pos - 100
    if pos < 0:
        pos = 100 + pos
    if pos == 0:
        count += 1
print("Result 1 : ",count)

# Part 2
pos = 50
count = 0
for line in data:
    if line[0] == "R":
        ctr = 100 - pos
        if int(line[1:]) >= ctr:
            count += ((int(line[1:]) - ctr) // 100 ) + 1
        pos += int(line[1:]) % 100
    else:
        ctr = pos
        if pos == 0:
            ctr = 100
        if int(line[1:]) >= ctr:
            count += ((int(line[1:]) - ctr) // 100) + 1
        pos -= int(line[1:]) % 100
    if pos > 99:
        pos = pos - 100
    if pos < 0:
        pos = 100 + pos
print("Result 2 : ",count)
```

```python
# Day 2
data = open('data/day2.txt').read().strip().split(',')
# Part 1
invalid_ids = []
for line in data:
    numbers = [int(x) for x in line.split('-')]
    min_num = min(numbers)
    max_num = max(numbers)
    for i in range(min_num, max_num + 1):
        s = str(i)
        l = len(s)
        if l % 2 == 0:
            mid = l // 2
            left = s[:mid]
            right = s[mid:]
            if left == right:
                invalid_ids.append(i)
print("Result 1 : ", sum(invalid_ids))

# Part 2
invalid_ids = []

def split_in_n_parts(s, n):
    part_len = len(s) // n
    return [s[i*part_len:(i+1)*part_len] for i in range(n)]

def all_parts_equal(parts):
    return all(p == parts[0] for p in parts)

for line in data:
    numbers = [int(x) for x in line.split('-')]
    min_num = min(numbers)
    max_num = max(numbers)
    for i in range(min_num, max_num + 1):
        s = str(i)
        l = len(s)
        for j in range(2, l+1):
            if l % j == 0:
                parts = split_in_n_parts(s, j)
                if all_parts_equal(parts):
                    invalid_ids.append(i)
                    break
print("Result 2 : ", sum(invalid_ids))
```

```python
# Day 3
data = open('data/day3.txt').read().strip().split('\n')
# Part 1
jolts = []
for line in data:
    start_dic = {}
    end_dic = {}
    for i in range(len(line)):
        if line[i] not in start_dic.keys():
            start_dic[line[i]] = i
    for i in range(len(line)):
            end_dic[line[i]] = i
    sorted_start = dict(sorted(start_dic.items(), reverse=True))
    sorted_end = dict(sorted(end_dic.items(), reverse=True))

    found = False
    for key, value in sorted_start.items():
        if found:
             break
        n1 = int(key)
        i = value
        for key, value in sorted_end.items():
             n2 = int(key)
             j = value
             if j > i:
                 jolts.append(n1*10 +n2)
                 found = True
                 break
             else:
                  continue
print("Result 1 : ", sum(jolts))

# Part 2
jolts = []
def find_next_number(line, start_index, remaining, jolt):
    start_dic = {}
    for i in range(start_index, len(line)):
        if line[i] not in start_dic.keys():
            start_dic[line[i]] = i
        sorted_start = dict(sorted(start_dic.items(), reverse=True))
    for key, value in sorted_start.items():
        next_number = key
        i = value
        if len(line) - i >= remaining:
            jolt += next_number
            return jolt, i+1
        else:
            continue
for line in data:
    remaining = 12
    jolt = ''
    start_index = 0
    while remaining > 0:
        jolt, start_index = find_next_number(
            line, start_index, remaining, jolt)
        remaining -= 1
    jolts.append(int(jolt))

print("Result 2 : ", sum(jolts))
```
